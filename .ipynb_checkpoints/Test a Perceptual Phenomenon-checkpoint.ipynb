(1) What is the independent variable? What is the dependent variable?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
df= pd.read_csv('stroopdata.csv')
df.head()
```

 >***The condition whether words are congruent and incongruent is independent but the  time taken to read is dependent.***

(2) What is an appropriate set of hypotheses for this task? Specify your null and alternative hypotheses, and clearly define any notation used. Justify your choices.

>***Null hypothesis
><br>H<sub>0</sub> : t<sub>congruent</sub> - t<sub>incongruent</sub> = 0
><br>The time taken by average population  for congruent and incongruent are equal.
<br>Alternative hypothesis
><br>H<sub>1</sub> : t<sub>congruent</sub> - t<sub>incongruent</sub>  != 0  
The time taken by average population is not equal for congruent and incongruent.***

>***We will use dependent sample t-test (two tailed) beacuse :
><br>This statistical procedure is used to determine whether the mean difference between two sets of observation is zero.
><br>We have less than 30 samples.
><br>We are assuming data is normally distributed.
><br>Because of our null and alternative hypothesis we are using two tailed.***

(4) Provide one or two visualizations that show the distribution of the sample data. Write one or two sentences noting what you observe about the plot or plots.

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))
sns.distplot(df['Congruent'] , color = 'red')
plt.xlabel('Time in seconds')
plt.title('Congruent time distribution')
plt.ylabel('Frequency');
```

>***As we can see in distribution plot congruent time is normally distributed .*** 

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))
sns.distplot(df['Incongruent'] , color = 'green')
plt.xlabel('Time in seconds')
plt.title('Incongruent time distribution')
plt.ylabel('Frequency');
```

>***This plot shows  incongruent time is right skewed may be because of outliers we can assume it as normally distributed for   t-test.***  

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))
sns.boxplot(data = df, order= ['Congruent','Incongruent'])
plt.ylabel('Time in seconds')
plt.title('Congruent & Incongruent box plot');
```


>*** From the above boxplot, we can see that the average completion time of Incongruent group i.e. 22.015917sec is higher than the average completion time of congruent group i.e. 14.051125sec .
><br>The box plot also shows that the incongruent group has two outliers. The distribution for both congruent and incongruent group looks normal.***

(5)  Now, perform the statistical test and report your results. What is your confidence level or Type I error associated with your test? What is your conclusion regarding the hypotheses you set up? Did the results match up with your expectations? **Hint:**  Think about what is being measured on each individual, and what statistic best captures how an individual reacts in each environment.

```python
# Perform the statistical test here
# https://docs.scipy.org/doc/scipy/reference/tutorial/stats.html
from scipy import stats
#print(df.shape)
degreesOfFreedom = 23
stats.t.ppf(0.95, degreesOfFreedom)

# Calculates the T-test on TWO RELATED samples of scores, a and b.
stats.ttest_rel(df['Congruent'], df['Incongruent'])
```

>***p value is under 0.05 therefore, we reject our null hypothesis. 
>Thus, we can conclude that incongruent and congruent has different time taken by average population complete tasks.***
